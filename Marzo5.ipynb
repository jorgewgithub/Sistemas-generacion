```python
# Definición de variables y parámetros
etha = 0.72
Q = 400 # m3/s - caudal
Hn = 175 # m - cabeza neta
n = 5  # número de conducciones

# Cálculo Potencia conducción
Po = etha*1000*9.8*Q*Hn
v = (2*9.8*Hn)**0.5
D = 0.52*Hn**(-1/7)*((Po/1000)/Hn)**(3/7) #Estimación diámetro Fahlbush

# Imprimir el resultado
print("Po = ", round(Po/1e6, 1), "MW ")
print("Potencia total de la central con", n, "unidades = ", round(n*Po/1e6, 1), "MW ")
print("velocidad agua = ", round(v, 1), "m/s ")
print("Diámetro estimado (Fahlbush) = ", round(D, 1), "m ")
```
